Hypothesis - T Test 

```python
import numpy as np
from scipy import stats
```

```python
sample_data = np.array([22, 25, 23, 27, 21, 24, 26, 28, 20, 29])
hypothesized_mean = 25
```

```python
t_statistic, p_value = stats.ttest_1samp(a=sample_data, popmean=hypothesized_mean)
```

```python
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05 # Significance level

    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the hypothesized population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the hypothesized population mean.")
```

Ho : mu = 70 (if pvalue > 0.05) : failed to reject Ho
Ha : mu != 70 (if pvalue < 0.05); reject Ho infavour of Ha
