```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
aq=data('airquality')
```

```python
aq.head()
```

```python
aq_long=pd.melt(frame=aq,id_vars=['Month','Day'],value_vars=['Ozone','Solar.R', 'Wind','Temp'],var_name='category')
aq_long
```

```python
aq_wide=aq_long.pivot_table(index=['Month','Day'], columns='category', values='value').reset_index()
```
